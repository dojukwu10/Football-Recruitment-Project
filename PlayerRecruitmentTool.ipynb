```python
from pyspark.ml.feature import VectorAssembler, StringIndexer
```

```python
archetypeEncoder = StringIndexer(inputCol='Playstyles', outputCol="Archetype").fit(df)
df = archetypeEncoder.transform(df)                                                          
```

```python
required_features = ['90s', 'Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgR', 'Sh', 'SoT', 'SoT%', 'G/Sh', 'G/SoT', 'Dist', 'npxG/Sh', 'G-xG', 'np:G-xG', 'TotDist', 'PrgDist', 'Cmp%S', 'Cmp%M', 'Cmp%L', 'xA', 'A-xAG', 'KP', 'F3rdPPA', 'PPA', 'CrsPA', 'PrgP', 'Live', 'Dead', 'FK', 'TB', 'Sw', 'Crs', 'TI', 'CK', 'In', 'Out', 'Str', 'Cmp', 'Off', 'BlockedPass', 'SCA', 'SCA90', 'PassLiveSCA', 'PassDeadSCA', 'TO', 'ShToS', 'Fld', 'Def', 'GCA', 'GCA90', 'PassLive', 'PassDead', 'TOSCA', 'ShSCA', 'FldSCA', 'DefSCA', 'Tkl', 'TklW', 'Def 3rd', 'Mid 3rd', 'Att 3rd', 'TklDef', 'Att', 'Tkl%', 'Lost', 'Blocks', 'ShBlocked', 'Pass', 'Int', 'Tkl+Int', 'Clr', 'Err', 'Touches', 'Def Pen', 'Def 3rdT', 'Mid 3rdT', 'Att 3rdT', 'Att Pen', 'LiveT', 'AttTO', 'Succ%', 'Tkld', 'Tkld%', 'Carries', 'TotDistC', 'PrgDistC', 'PrgC', 'F3Carry', 'CPA', 'Mis', 'Dis', 'Rec', 'CrdY', 'CrdR', '2CrdY', 'Fls', 'PKwon', 'PKcon', 'OG', 'Recov', 'ADWon', 'ADLost', 'Archetype']
```

```python
vec_assembler = VectorAssembler(inputCols=required_features, outputCol='features')
```

```python
vec_df = vec_assembler.transform(df)
```

```python
print(vec_df.columns)
```

```python
vec_df.show(2)
```

```python
columns_to_drop = ['Playstyles']
vec_df = vec_df.drop(*columns_to_drop)
```

```python
vec_df.show(2)
```

```python
train_df, test_df = vec_df.randomSplit([0.7,0.3])
```

```python
train_df.count()
```

```python
test_df.count()
```

```python
from pyspark.ml.classification import LogisticRegression,DecisionTreeClassifier
```

```python
lr = LogisticRegression(featuresCol='features', labelCol='Archetype')
```

```python
lr_model = lr.fit(train_df)
```

```python
y_pred = lr_model.transform(test_df)
```

```python
y_pred.show()
```

```python
y_pred.select('Names', 'probability', 'prediction').show(100)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
multi_evaluator = MulticlassClassificationEvaluator(labelCol='Archetype') 
```

```python
multi_evaluator.evaluate(y_pred)
```
